```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("F:\Gitworkplace\Data_science_deployment\CarPricePrediction\car data.csv")
```

```python
df.head()
```

Distribution of Car Prices

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Selling_Price'], kde=True)
plt.title('Distribution of Selling Prices')
plt.xlabel('Selling Price (in lakhs)')
plt.ylabel('Frequency')
plt.show()
```

Number of Cars by Fuel Type

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Fuel_Type', data=df)
plt.title('Number of Cars by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Cars')
plt.show()
```

Number of Cars by Transmission Type

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Fuel_Type', data=df)
plt.title('Number of Cars by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Cars')
plt.show()
```

Number of Cars by Transmission Type

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Transmission', data=df)
plt.title('Number of Cars by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Number of Cars')
plt.show()
```

Selling Price vs. Present Price

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

# Define colors based on Fuel_Type
colors = {'Petrol': 'red', 'Diesel': 'blue', 'CNG': 'green'}

# Map Fuel_Type to colors
df['Color'] = df['Fuel_Type'].map(colors)

# Plot scatter plot with color mapping
sns.scatterplot(x='Present_Price', y='Selling_Price', hue='Fuel_Type', data=df, palette=colors.values())
# Add legend
plt.legend(title='Fuel Type')
plt.title('Selling Price vs. Present Price by Fuel Type')
plt.xlabel('Present Price (in lakhs)')
plt.ylabel('Selling Price (in lakhs)')
plt.show()
```
